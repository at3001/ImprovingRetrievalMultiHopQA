```python
def generate_negative_pairs(df):
  res_df = []
  for x in range(len(df)):
    sf = df.iloc[x].supporting_facts['title']
    titles = df.iloc[x].context['title']
    sentences = df.iloc[x].context['sentences']

    n = len(titles)
    for i in range(n):
      if titles[i] not in sf:
        res_df.append([df.iloc[x].question, titles[i], sentences[i], 0])
      else:
        res_df.append([df.iloc[x].question, titles[i], sentences[i], 1])
  return res_df

```

```python
def convert_list_to_sent(para_list):
  """
  Para list is a list of list of sentences
  """

  return [' '.join(x) for x in para_list]
```

```python
contexts = []
for i in tqdm(range(3000)):
  contexts += hpqa[i]['context']['sentences']

contexts = [tuple(context) for context in contexts]
contexts = list(set(contexts))
contexts = [list(context) for context in contexts]
contexts = convert_list_to_sent(contexts)
print(contexts[0])
```

```python
questions = [hpqa[i]['question'] for i in range(200)]
supporting_titles = [hpqa[i]['supporting_facts']['title'] for i in range(100)]
answers = [hpqa[i]['answer'] for i in range(200)]
level = [hpqa[i]['level'] for i in range(200)]

supporting_articles = []
for i in tqdm(range(100)):
  sup_titles = list(set(supporting_titles[i]))
  sup_1 = sup_titles[0]
  sup_2 = sup_titles[1]

  idx_1 = hpqa[i]['context']['title'].index(sup_1)
  idx_2 = hpqa[i]['context']['title'].index(sup_2)

  supporting_articles.append([' '.join(hpqa[i]['context']['sentences'][idx_1]), \
                              ' '.join(hpqa[i]['context']['sentences'][idx_2])])
```

```python
from llama_index.core import VectorStoreIndex
from llama_index.core import Document

documents = [Document(text=t) for t in contexts]
index = VectorStoreIndex.from_documents(
    documents,
    top_k = 5,
    llm = OpenAI(temperature = 0, model = 'gpt-3.5-turbo')
)
```

```python

similarity_top_k = 5
hit_per_q = []
num_hit = 0
num_hit_rel = 0
responses = []
mrr_sum = 0

for i in tqdm(range(len(questions))):
  hit_perc = 0
  q = questions[i]
  hit_flag = False

  engine = index.as_query_engine(similarity_top_k = similarity_top_k)
  query_result = engine.query(q)

  hits = [query_result.source_nodes[i].text for i in range(len(query_result.source_nodes))]

  responses.append(query_result.response)

  for j in range(len(supporting_articles[i])):
    if supporting_articles[i][j] in hits:
      hit_flag = True
      num_hit += 1
      hit_perc += 1

  for j in range(len(hits)):
    if hits[j] in supporting_articles[i]:
      mrr_sum += (1/(j+1))
      break


  if hit_flag:
    num_hit_rel += 1
  hit_per_q += [hit_perc]
```

```python
def get_text_embedding(text):
  tosubmit = text.replace("\n", " ")
  response = openai_client.embeddings.create(input=[tosubmit], model = EMBED_MODEL)
  return response.data[0].embedding

def construct_dataset(queries, ranked_docs):

  # Columns - query, document, relvance score, rank
  # Maybe store them as embeddings within this dataframe?

  d = {
      'queries': queries.copy(),
      'documents': [x[1] for x in ranked_docs],
      'rank': [x[0] for x in ranked_docs],
      'relevance_score': [x[2] for x in ranked_docs],
  }

  df = pd.DataFrame(data = d)
  df[EMBED_COLUMNS[0]] = df['queries'].apply(get_text_embedding)
  df[EMBED_COLUMNS[1]] = df['document'].apply(get_text_embedding)
  df.to_csv("Experiment 1 Raw Embedding Data")
  return df
```

```python
def get_ranked_docs(query):
  engine = index.as_query_engine(similarity_top_k = 10)
  query_result = engine.query(query)
  hits = [query_result.source_nodes[i].text for i in range(len(query_result.source_nodes))]

  results = co.rerank(query = query, documents = hits, top_n = 3, model = RERANK_MODEL, return_documents = True)

  out = []

  for idx, r in enumerate(results.results):
    out.append((idx + 1, r.document.text, r.relevance_score))
  return out

```

```python
def mse_loss(predicted, actual):
  diff = predicted - actual
  return torch.sum(diff*diff) / diff.numel()
```

```python
class EmbeddingDataset(torch.utils.data.Dataset):
  def __init__(self, df):
    self.df = df

  def __len__(self):
    return len(self.df)
  def __getitem__(self, idx):
    qembed = self.df["QUERY_EMBEDDINGS"][idx]
    dembed = self.df["DOCUMENT_EMBEDDINGS"][idx]
    idx = self.df["index"][idx]
    score = self.df["relevance_score"][idx]
    return (qembed, dembed, idx, score)
```

```python
ed = EmbeddingDataset(qddf)
```

```python
train_loader = torch.utils.data.DataLoader(ed, batch_size = 1, shuffle = True)
```

```python
# Testing Train Loader
matrix = torch.randn(len(test_q_embed), len(test_d_embed), requires_grad = True).float()
epochs = 5
learning_rate = 1e-4
for _ in tqdm(range(epochs)):
  for qembed, dembed, idx, score in train_loader:
    predicted_relevance = torch.nn.functional.softmax(torch.tensor(qembed).float() @ matrix @ torch.tensor(dembed).float())
    loss = mse_loss(predicted_relevance, score)
    loss.backward()
    with torch.no_grad():
      matrix -= matrix.grad * learning_rate
      matrix.grad.zero_()



```
